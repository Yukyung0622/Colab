```python
"""
날짜 : 2022/04/28
이름 : 김유경
내용 : Pandas 연습문제1 탐색분석
"""
#라이브러리 선언
import pandas as pd
```

```python
#1-1.데이터 불러오기
data_url = 'https://raw.githubusercontent.com/chhak/dataset/master/chipo.csv'
df = pd.read_csv(data_url)
df
```

```python
#1-2.quantity 컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하시오.
df.loc[df['quantity']==3].head()
```

```python
#1-3.quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하시오.
df.loc[df['quantity']==3].head().reset_index(drop=True)
```

```python
#1-4.quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하시오.
new_df = df[['quantity', 'item_price']]
new_df
```

```python
#1-5.item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하시오.
df['new_price'] = df['item_price'].str[1:].astype('float')
result = df['new_price'].head()
result
```

```python
#1-6.new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하시오.
result = len(df.loc[df.new_price <=5])
result
```

```python
#1-7.item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하고 index 값을 초기화 하여 출력하시오.
df.loc[df.item_name == 'Chicken Salad Bowl'].reset_index(drop=True)
```

```python
#1-8.new_price값이 9 이하 item_name 값이 Chicken Salad Bowl 인 데이터 프레임에서 상위 5개만 출력하시오.
df.loc[(df.item_name == 'Chicken Salad Bowl') & (df.new_price <= 9)].head(5)
```

```python
#1-9.df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여 상위 3개만 출력하시오
df.sort_values('new_price').reset_index(drop=True).head(3)
```

```python
#1-10.df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하시오.
df.loc[df.item_name.str.contains('Chips')]
```

```python
#1-11.df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하시오.
df.iloc[:, ::2]
```

```python
#1-12.df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하시오.
df.sort_values('new_price',ascending=False).reset_index()
```

```python
#1-13.df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱 하시오.
df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')]
```

```python
#1-14.df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터프레임으로 생성 후, 
#item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남기시오. 
df2 = df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')]
df3 = df2.drop_duplicates('item_name')
df3
```

```python
#1-15.df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터프레임으로 생성 후, 
#item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남기시오.
df2 = df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')]
df3 = df2.drop_duplicates('item_name', keep='last')
df3
```

```python
#1-16.df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱 하시오.
df.loc[df.new_price >= df.new_price.mean()]
```

```python
#1-17.df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하시오
df.loc[df.item_name == 'Izze', 'item_name'] = 'Fizzy Lizzy'
df
```

```python
#1-18.df의 데이터 중 choice_description 값이 NaN 인 데이터의 갯수를 구하시오
df.choice_description.isnull().sum()
```

```python
#1-19.df의 데이터 중 choice_description 값이 NaN 인 데이터를 NoData 값으로 대체하시오.(loc 이용)
df.loc[df.choice_description.isnull(), 'choice_description'] = 'NoData'
df
```

```python
#1-20.df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 상위 3개만 인덱싱 하시오
df[df.choice_description.str.contains('Black')].head(3)
```

```python
#1-21.df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력하시오.
len(df.loc[~df.choice_description.str.contains('Vegetables')])
```

```python
#1-22.df의 데이터 중 item_name 값이 N으로 시작하는 데이터 상위 3개만 출력하시오.
df[df.item_name.str.startswith('N')].head(3)
```

```python
#1-23.df의 데이터 중 item_name 값의 단어갯수가 30개 이상인 데이터를 인덱싱하시오.
df[df.item_name.str.len() >= 30]
```

```python
#1-24.df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수를 출력하시오.
lst = [1, 69, 2.39, 3.39, 4.45]
df.loc[df.new_price.isin(lst)]
```
