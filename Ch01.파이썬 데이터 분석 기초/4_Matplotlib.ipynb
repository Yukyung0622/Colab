```python
"""
날짜 : 2022/04/26
이름 : 김유경
내용 : 파이썬 Matplotlib실습하기

Matplotlib
 - 다양한 시각화 그래프를 지원하는 파이썬 시각화 라이브러리
 - 데이터 탐색 또는 분석결과를 시각화하기 위해 사용하는 파이썬 라이브러리
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##################################
#라인차트
plt.title('line')
plt.plot([1, 3, 4, 8, 10, 17])
plt.show()
```

```python
#x축 data
plt.title('line2')
plt.plot([10, 20, 30, 40, 50, 60],[1, 3, 4, 8, 10, 17])
plt.show()
```

```python
#스타일
plt.title('line3')
plt.plot([10, 20, 30, 40, 50, 60],[1, 3, 4, 8, 10, 17], 'ys:')
plt.grid(True)
plt.show()
```

```python
#멀티 라인차트
values = np.arange(0, 5, 0.2)
print(values)

plt.title('line4')
plt.plot(values, values, 'g--')
plt.plot(values, values ** 2, 'b--')
plt.plot(values, values ** 3, 'k--')
plt.plot(values, values ** 4, 'r--')
plt.show()
```

```python
#라인차트 연습문제
flights = sns.load_dataset('flights')
flights

x = flights.drop_duplicates('year') ['year']
y = flights.groupby(by='year') ['passengers'].agg('sum')

print('x :', x)
print('y :', y)

plt.title('flights')
plt.plot(x,y)
plt.show()
```

```python
#산점도 연습문제
tips = sns.load_dataset('tips')
tips

x = tips['total_bill']
y = tips['tip']

plt.title('Tips')
plt.scatter(x, y)
plt.xlabel('Total Bill')
plt.xlabel('Tip')
plt.show()
```

```python
##################################
#박스상자
data1 = np.random.randint(100, size=5)
data2 = np.random.randint(100, size=5)

print('data1 : ', data1)
print('data2 : ', data2)

plt.boxplot([data1, data2])
plt.xticks([1, 2], ['data1', 'data2'])
plt.show()
```

```python
#박스상자 연습문제
iris = sns.load_dataset('iris')
iris

dataset1 = iris[iris['species'] == 'setosa']['sepal_length']
dataset2 = iris[iris['species'] == 'versicolor']['sepal_length']
dataset3 = iris[iris['species'] == 'virginica']['sepal_length']

plt.title('iris boxplot')
plt.boxplot([dataset1, dataset2, dataset3], labels=['setosa', 'versicolor', 'virginica'])
plt.show()
```

```python
#3차원 산점도
import plotly.express as px

fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='sepal_width', color='species')
fig.show()
```
